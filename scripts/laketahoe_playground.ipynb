```python
# 📂 5. ファイル管理ユーティリティ
def show_files():
    """作成されたファイル一覧を表示"""
    files = [f for f in os.listdir('.') if f.endswith('.py')]
    print("📂 現在のPythonファイル:")
    for f in files:
        size = os.path.getsize(f)
        print(f"  📄 {f} ({size} bytes)")

def edit_file(filename, line_number, new_content):
    """特定の行を編集する関数"""
    try:
        with open(filename, 'r') as f:
            lines = f.readlines()
        
        if 1 <= line_number <= len(lines):
            lines[line_number - 1] = new_content + '\n'
            
            with open(filename, 'w') as f:
                f.writelines(lines)
            
            print(f"✅ {filename} の {line_number} 行目を編集しました")
        else:
            print(f"❌ 行番号 {line_number} は範囲外です (1-{len(lines)})")
    except Exception as e:
        print(f"❌ エラー: {e}")

def backup_file(filename):
    """ファイルのバックアップを作成"""
    import shutil
    backup_name = f"{filename}.backup"
    shutil.copy(filename, backup_name)
    print(f"✅ {filename} を {backup_name} としてバックアップしました")

# ファイル一覧を表示
show_files()

# ファイル編集のデモ
print("\n🔧 ファイル編集デモ:")
backup_file('experiment_config.py')
edit_file('experiment_config.py', 3, 'BATCH_SIZE = 128  # 修正されました!')

# 編集結果を確認
with open('experiment_config.py', 'r') as f:
    lines = f.readlines()
print("\n📖 編集後の内容:")
for i, line in enumerate(lines[:10], 1):
    print(f"  {i}: {line.rstrip()}")
```
